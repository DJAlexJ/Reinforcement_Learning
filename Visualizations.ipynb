# Baseline Reinforce VS Simple Reinforce

Baseline function was learnt on 3 trajectories and then both reinforce algorithms used 3 trajectoris

```python
rein = pd.read_csv('D:/Reinforcement_Learning/df_reinforce_2_trajectories_2000.csv')
rein_basel = pd.read_csv('D:/Reinforcement_Learning/df_reinforce_2_trajectories_baseline_2000.csv')
rein_basel_retrain = pd.read_csv('D:/Reinforcement_Learning/df_reinforce_2_trajectories_baseline_2000_retrain.csv')
```

```python
r_reward = rein.rewards
r_mean_reward = rein.mean10_rewards
rb_reward = rein_basel.rewards
rb_mean_reward = rein_basel.mean10_rewards
rbr_reward = rein_basel_retrain.rewards
rbr_mean_reward = rein_basel_retrain.mean10_rewards
```

```python
fg = plt.figure(figsize = (12, 8), constrained_layout = True)
grids = fg.add_gridspec(2, 2)
fig_ax_1 = fg.add_subplot(grids[0, 0])
plt.plot(r_reward)
plt.plot(r_mean_reward)
plt.title('Reinforce without baseline', fontsize = 17)
plt.xlabel('Episode', fontsize = 15)
plt.ylabel('Reward', fontsize = 15)
plt.text(500, 5, 'varince = {}'.format(np.var(r_reward)), fontsize = 13)
fig_ax_2 = fg.add_subplot(grids[0, 1])
plt.plot(rb_reward)
plt.plot(rb_mean_reward)
plt.title('Reinforce with baseline', fontsize = 17)
plt.xlabel('Episode', fontsize = 15)
plt.ylabel('Reward', fontsize = 15)
plt.text(500, 5, 'varince = {}'.format(np.var(rb_reward)), fontsize = 13)
fig_ax_3 = fg.add_subplot(grids[1, :])
plt.plot(r_mean_reward)
plt.plot(rb_mean_reward)
plt.title('Mean rewards', fontsize = 17)
plt.xlabel('Episode', fontsize = 15)
plt.ylabel('Reward', fontsize = 15)
plt.legend(['Without baseline', 'With baseline'])
```
